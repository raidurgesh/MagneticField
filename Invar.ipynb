```python
import numpy as np
import matplotlib.pyplot as plt

# calculation of # atoms in cm, g
Rad=4.8/2;
Thick=1;
Den=8.05;
AtNum=0.64*55.847+0.36*58.70
Na=6.022e23;
Vol=np.pi*Rad**2*1
NumAt=Vol*Den*Na/AtNum
# with number of atoms calculated, we calcualte the total magnetic moment (assuming 1.6 Bohr magnetron, units in Am^2):
BU=9.27e-24;
Moment=NumAt*1.6*BU
mu=4.*np.pi*1e-7
x_TM = np.linspace(0.1, 1.25, 25)
y_TM = np.linspace(0.1, 1.25, 25)
# distance in cm
dist_TM = np.sqrt(x_TM**2+y_TM**2)*100
# we calculate the feild using, B=mu*moment/2*pi*R^3, in nT
field=mu*Moment/(2.*np.pi*(dist_TM/100)**3)*1e9
field_1m=mu*Moment/(2.*np.pi*0.6**3)*1e9
print("distance to TM (cm) : ", dist_TM)
print("field (nT) : ", field)
print("Summary : ")
print("Volume  (cc) = ", Vol)
print("Moment (Ampere*m^2) = ", Moment)
print("Permeability (Henry/m) : ", mu)
print("Magnetic Field at 1m (nT) = ", field_1m)
#print("Volume  (cc),  Moment (Ampere*m^2), Permeability (Henry/m), Magnetic Field at 1m (nT)")
#print(Vol, Moment, mu, field_1m)
plt.style.use('bmh')
plt.plot(dist_TM, field, 'g-')
plt.plot(dist_TM,np.ones(np.shape(dist_TM))*500,'r--')
plt.yscale("log")
#plt.xscale("log")
plt.title("Field vs Distance to TM for a single Invar Disc")
plt.xlabel("Radial Distance to TM (cm)")
plt.ylabel("Magnetic Field (nT)")
plt.show()
```
