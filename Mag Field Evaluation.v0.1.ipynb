```python
import numpy as np
import matplotlib.pyplot as plt

# Moments from IRD2/PRD for respective CFIs
Moment_CFI = 50e-3
Moment_LH = 125e-3
Moment_PMON = 1e-3
Moment_FOH = 0.025e-3

#Constants
mu=4.*np.pi*1e-7

# distance in m
x_TM = np.linspace(1e-6, 1.25, 1000)
y_TM = np.linspace(1e-6, 1.25, 1000)
dist_TM = np.sqrt(x_TM**2+y_TM**2)

# distance to CFI in m (from co-engineering and ESA/Brian)
dist_TM_LH = 1.2185 
dist_TM_FRS_E = 1.0582
dist_TM_FRS_O = 1.1102 
dist_TM_PMON = 0.1428 
dist_TM_CMD = 1.46025
dist_TM_FOH = 0.08

# we calculate the feild using, B=mu*moment/2*pi*R^3, in nT
field_CFI = mu*Moment_CFI/(2.*np.pi*(dist_TM)**3)*1e9
field_LH = mu*Moment_LH/(2.*np.pi*(dist_TM)**3)*1e9
field_PMON = mu*Moment_PMON/(2.*np.pi*(dist_TM)**3)*1e9
field_FOH = mu*Moment_FOH/(2.*np.pi*(dist_TM)**3)*1e9

#Calculate magnetic field at minimum distance for reference
field_LH_1218=mu*Moment_LH/(2.*np.pi*(dist_TM_LH)**3)*1e9
field_FRS_E_1058=mu*Moment_CFI/(2.*np.pi*(dist_TM_FRS_E)**3)*1e9
field_FRS_O_1110=mu*Moment_CFI/(2.*np.pi*(dist_TM_FRS_O)**3)*1e9
field_PMON_142=mu*Moment_PMON/(2.*np.pi*(dist_TM_PMON)**3)*1e9
field_CMD_1460=mu*Moment_CFI/(2.*np.pi*(dist_TM_CMD)**3)*1e9
field_FOH_80=mu*Moment_FOH/(2.*np.pi*(dist_TM_FOH)**3)*1e9

#Printout Summary
print("Magnetic Field at the TM position due to LH (nT) = ", field_LH_1218)
print("Magnetic Field at the TM position due to FRS-E (nT) = ", field_FRS_E_1058)
print("Magnetic Field at the TM position due to FRS-O (nT) = ", field_FRS_O_1110)
print("Magnetic Field at the TM position due to PMON (nT) = ", field_PMON_142)
print("Magnetic Field at the TM position due to CMD (nT) = ", field_CMD_1460)
print("Magnetic Field at the TM position due to FOH (nT) = ", field_FOH_80)

#Plot
plt.figure(figsize=(10,8))
plt.plot(dist_TM, field_CFI, 'b-', label = "FRS-E, FRS-O, CMD", linewidth=1.0)
plt.plot(dist_TM, field_LH, 'r-', label = "LH", linewidth=1.0)
plt.plot(dist_TM, field_PMON, 'g-', label = "PMON", linewidth=1.0)
plt.plot(dist_TM, field_FOH, 'm-', label = "FOH", linewidth=1.0)

# Horizontal Lines on Plot
plt.plot(dist_TM,np.ones(np.shape(dist_TM))*0.3,  'y-', label = "Detection Limit", linewidth=1)
plt.plot(dist_TM,np.ones(np.shape(dist_TM))*1, 'c-', label = "Measurement Limit", linewidth=1)

# Vertical Lines on Plot 
plt.axvline(x = 0.08, color="m", label = "FOH Distance", linestyle="--", linewidth=1)
plt.axvline(x = 0.1428, color="g", label = "PMON Distance",linestyle="--", linewidth=1)
plt.axvline(x = 1.0582, color="b", label = "FRS-E Distance",linestyle="--", linewidth=1)
plt.axvline(x = 1.1102, color="b", label = "FRS-O Distance",linestyle=":", linewidth=1)
plt.axvline(x = 1.2185, color="r", label = "LH Distance",linestyle="--", linewidth=1)
plt.axvline(x = 1.460255, color="b", label = "CMD Distance", linestyle="-.", linewidth=1)

plt.yscale("log")
#plt.xscale("log")
plt.xlim(1e-5, 1.5)
plt.ylim(1e-3, 1e6)
plt.title("Mag Field vs Distance to TM from an ideal dipole along dipole axis")
plt.xlabel("Radial Distance to TM (cm)")
plt.ylabel("Magnetic Field (nT)")

plt.legend()
plt.grid(color = 'k', linestyle = '-', linewidth = 0.5)
plt.show()
```
